```python
import numpy as np
import pandas as pd
```

```python
fi = open('model/q_table.csv')
```

```python
lines = fi.readlines()
```

```python
lines
```

```python
q_table = pd.DataFrame(columns=[0,1,2,3], dtype=np.float64)
```

```python
lines = lines[1:]
for line in lines:
    line = line.strip().split(' ')
    line = [float(item) for item in line if (item != '')]
    state = int(line[0])
    xxx = line[1:]
    print((state, xxx))
    q_table = q_table.append(pd.Series(xxx, index=q_table.columns, name=state))
```

```python
q_table.to_csv('out.csv')
```
